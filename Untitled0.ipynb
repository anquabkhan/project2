```python
from keras.models import Sequential
```

```python
from keras.layers import Conv2D, MaxPool2D, Dropout, MaxPooling2D, Flatten, Dense 
```

```python
from keras.optimizers import Adam 
```

```python
cnn_model5 = Sequential()
```

```python
cnn_model5.add(Conv2D(filters= 64, kernel_size=(3,3), activation='relu', input_shape= Input_shape ))
```

```python
cnn_model5.add(Conv2D(64, kernel_size=(3,3), activation='relu'))
```

```python
cnn_model5.add(MaxPooling2D(2,2))
```

```python
cnn_model5.add(Dropout(0.4))
```

```python
cnn_model5.add(Conv2D(filters= 128, kernel_size=(3,3), activation='relu')) 
```

```python
cnn_model5.add(Conv2D(filters=128, activation='relu', kernel_size=(3,3)))
```

```python
cnn_model5.add(MaxPooling2D(2,2))
```

```python
cnn_model5.add(Flatten())
```

```python
cnn_model5.add(Dense(units= 1024, activation= 'relu'  ))                                  
```

```python
cnn_model5.add(Dense(units= 1024, activation= 'relu'))
```

```python
cnn_model5.add(Dense(units= 100, activation= 'softmax')) 
```

```python
cnn_model5.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])  
```

```python
history = cnn_model5.fit(X_train2, y_train2, epochs= 1, batch_size= 32 , shuffle= True)                                                     
```
